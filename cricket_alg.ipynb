```python
#imported libraries
import pandas as pd
import numpy as np
import csv 
import matplotlib as mp
import matplotlib.pyplot as plt
import math
```

```python
#Data Imporation and Initial Organization
dataframe = pd.read_csv('most_runs_in_cricket.csv')
def career_length(career):  # 1111-2222
    return int(career[-4:]) - int(career[:4])
runs = dataframe['Runs'] #Y1
sixes = dataframe['6s'] #Y2
fours = dataframe['4s'] #Y3
notout = dataframe['NO'] #Y4
careers = dataframe['Span']
length_careers = [] #X
for i in careers:
    length_careers.append(career_length(i))

distScores = [[] for i in range(len(length_careers))]

print(distScores)
```

```python
# Distance Calculation
def calcDist(x, y, a, b):
    a =  -a
    c = -b
    d = (a*x + y + c)/math.sqrt(a**2 + 1)
    return d

def distList(arr,a,b):
    for i in range(len(length_careers)):
        d = calcDist(length_careers[i],arr[i],a,b)
        distScores[i].append(d)
```

```python
#Runs
xMAX = max(length_careers)
yMAX = max(runs)
figure, axis = plt.subplots(1,2)

axis[0].plot(length_careers, runs, 'bo') # ([x values], [y values], 'color')
axis[0].axis([0, 1.1*xMAX, 0, 1.1*yMAX])

a, b = np.polyfit(length_careers, runs, 1)
lc = np.array(length_careers)
axis[0].plot(lc, a*lc+b, color = 'red')


temp = [i[1]-i[0] for i in list(zip((a*lc+b),runs))]
resid = np.array(temp)
rMAX = max(resid)
rMIN = min(resid)
axis[1].scatter(lc,resid, color = 'blue')
axis[1].axis([0, 1.1*xMAX, 1.1*rMIN, 1.1*rMAX])
axis[1].axhline(y=0, color = 'red')

axis[0].set_title("Runs vs. Career Length")
axis[1].set_title("Runs vs. Career Length (Residual Plot)")

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
print("Runs: y = " + str(a) + "x + " + str(b))

corr_matrix = np.corrcoef(a*lc+b, runs)
corr = corr_matrix[0,1]
R_sq = corr**2
print("Coefficient of Correlation (R): " + str(corr))
print("Coefficient of Determination (R^2): " + str(R_sq))

distList(runs,a,b)
```

```python
distScoresSum = [sum(i) for i in distScores]
maxGrade = max(distScoresSum)
minGrade = min(distScoresSum)
print(maxGrade, minGrade)
gradingDict = {"A+": maxGrade, "A": maxGrade-(maxGrade-minGrade)/10, "A-": maxGrade-2*(maxGrade-minGrade)/10, "B+": 
    maxGrade-3*(maxGrade-minGrade)/10, "B": maxGrade-4*(maxGrade-minGrade)/10, "B-": maxGrade-5*(maxGrade-minGrade)/10, 
    "C+": maxGrade-6*(maxGrade-minGrade)/10, "C": maxGrade-7*(maxGrade-minGrade)/10, "C-": maxGrade-8*(maxGrade-minGrade)/10,
    "D+": maxGrade-9*(maxGrade-minGrade)/10, "D": minGrade}

names = dataframe['Player']
final = []

for i in range(len(distScoresSum)):
    curGrade = ""
    for j in gradingDict.keys():
        if distScoresSum[i] >= gradingDict[j]:
            curGrade = j
            break
    final.append([names[i],curGrade,i])
    print(names[i] + "\t" + "\t" + curGrade + "\t" + str(i+1))
```
